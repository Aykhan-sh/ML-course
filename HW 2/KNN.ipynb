```python
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import statsmodels.api as sm
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
import matplotlib
```

```python
X, y = datasets.load_iris(return_X_y=True, as_frame=True)
X
```

# Quick EDA

```python
fig, ax = plt.subplots(nrows=4, ncols=2, figsize=(23,23))
for idx, i in enumerate(X.columns):
    ax[idx][0].set_title(i)
    sm.qqplot(X[i], line='s', ax=ax[idx][0])
    ax[idx][1].set_title(i)
    sns.histplot(X[i], ax=ax[idx][1])
```

```python
plt.figure(figsize=(13,5))
plt.title("Class distribuion")
sns.countplot(x=y)
```

# KNN

```python
X = (X-X.mean())/X.std()  # normalizing data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23, stratify=y)  # splitting data
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
preds = model.predict(X_test)
accuracy_score(y_test, preds)
```
